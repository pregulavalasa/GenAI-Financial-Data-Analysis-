### Read Transaction Data

```python
# Read Data
import pandas as pd

df = pd.read_csv("Amex_Transactions.csv")
df.head()
```

```python
unq_det = df["Extended Details"].unique()
len(unq_det)
```

```python
unq_det
```

```python
unq_cat = []
for i in unq_det:
    j = i.split('\n')
    try:
        j = j[1]
        unq_cat.append(j)
    except:
        unq_cat.append(i)

unq_cat = pd.DataFrame({'cat':unq_cat})
unq_cat = unq_cat['cat'].unique()
unq_cat
```

### Categorise Credit Transactions with Llama3.2

```python
response = llm.invoke(" Can you add an appropriate category next to each of the following expenses. Respond with a list of expenses and categories. For example, AplPay BT*DD *DOORDASH BIGLANTER - Restaurant , ENTERPRISE RENT-A-CAR - Travel, etc.  Here are the expenses:" + ", ".join(unq_cat[0:15]))
response.split('\n')
```

```python
# Index list
def hop(start, stop, step):
    for i in range(start, stop, step):
        yield i
    yield stop

index_list = list(hop(0, len(unq_cat), 15))
index_list
```

```python
def cat_trans (trans, llm):
    response = llm.invoke("Can you add an appropriate category next to each of the following expenses. Respond with a list of expenses and categories. For example, AplPay BT*DD *DOORDASH BIGLANTER - Restaurant , ENTERPRISE RENT-A-CAR - Travel, etc.  Here are the expenses:" + trans)
    response = response.split('\n')

    print(response)

    trans_cat = pd.DataFrame({'Transactions Vs Category': response})
    trans_cat[['Transaction', 'Category']] = trans_cat['Transactions Vs Category'].str.split(' - ', n=1, expand = True)

    return trans_cat
```

```python
#Testing

cat_trans('AplPay SPICE WORLD MARKET, AplPay AROMA INDIAN & NEPALE, AplPay POETS BILLIARDS, Amazon Web Services', llm)
```

```python
# Intialise the categories_df_all dataframe
categories_df_all = pd.DataFrame()

# Loop through the index_list
for i in range(0, len(index_list)-1):
    transaction_names = unq_cat[index_list[i]:index_list[i+1]]
    transaction_names = ', '.join(transaction_names)

    categories_df = cat_trans(transaction_names, llm)
    categories_df_all = pd.concat([categories_df_all, categories_df], ignore_index=True)
```

```python
categories_df_all
```

```python
# Get unique categories in categories_df_all
unique_categories = categories_df_all["Category"].unique()
unique_categories
```

```python
# Drop NA values
categories_df_all = categories_df_all.dropna()

# If category contains "Gas", then categorise as "Gas"
categories_df_all.loc[categories_df_all['Category'].str.contains("Gas"), 'Category'] = "Gas"
# If category contains "Groceries", then categorise as "Groceries"
categories_df_all.loc[categories_df_all['Category'].str.contains("Groceries|Grocery"), 'Category'] = "Groceries"
# If category contains "Restaurant" or "Delivery" or "Drinks", then categorise as "Restaurants or Take Out"
categories_df_all.loc[categories_df_all['Category'].str.contains("Restaurant|Drinks|Delivery"), 'Category'] = "Restaurants or Take Out"
# If category contains "Clothing", then categorise as "Clothing"
categories_df_all.loc[categories_df_all['Category'].str.contains("Clothing"), 'Category'] = "Clothing"
# If category contains "Entertainment", then categorise as "Entertainment"
categories_df_all.loc[categories_df_all['Category'].str.contains("Entertainment"), 'Category'] = "Entertainment"
# If category contains "Health" or "Retail", then categorise as "Retail"
categories_df_all.loc[categories_df_all['Category'].str.contains("Retail|Store"), 'Category'] = "Retail"
# If category contains "Transportation" or "Taxi" or "Travel", then categorise as "Transportation"
categories_df_all.loc[categories_df_all['Category'].str.contains("Travel|Taxi|Transportation"), 'Category'] = "Transportation"
```

```python
# Remove the numbering eg "1. " from Transaction column
categories_df_all['Transaction'] = categories_df_all['Transaction'].str.replace(r'\d+\.\s+', '')
categories_df_all
```

```python
# Merge the categories_df_all with the original transactions dataframe (df)
df = pd.read_csv("Amex_Transactions.csv")
df['trans'] = df['Extended Details'].str.split('\n')

df['trans_u'] = [i[1] if len(i)>1 else i[0] for i in df['trans']]
df['trans_u']
```

```python
df = pd.merge(df, categories_df_all, left_on='trans_u', right_on='Transaction', how='left')
df
```

```python
df.to_csv("Amex_Transactions_categorized.csv", index=False)
```
